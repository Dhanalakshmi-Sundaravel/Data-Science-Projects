## Question 2:

```python
max(beer_data['review_overall'])
```

```python
overall = beer_data['review_overall'] == 5.0
```

```python
aroma = beer_data['review_aroma'] == 5.0
```

```python
appearance = beer_data['review_appearance'] == 5.0
```

```python
palate = beer_data['review_palate'] == 5.0 
```

```python
taste = beer_data['review_taste'] == 5.0
```

```python
best_rated = beer_data[overall & aroma & appearance & palate & taste]
```

```python
best_rated_sorted = (best_rated.groupby('beer_beerid')['beer_beerid'].count())
```

```python
best_rated_sorted = pd.DataFrame({'beerid':best_rated_sorted.index, 'count':best_rated_sorted.values})
```

```python
best_rated_sorted.sort_values('count',ascending = False)
```
